문제 1 답) GAN은 반대되는 목적을 가진 두 모듈 Generator와 discriminator이 대결하는 구조를 통해 이미지 생성을 학습합니다. Generator는 실제 데이터를 학습해 비슷한 거짓 데이터를 생성해내며 최대한 실제 데이터와 비슷한 거짓 데이터를 생성하는 것이 목적입니다. Discriminator는 입력으로 받은 데이터가 실제 데이터인지 거짓 데이터인지 판별하며 실제 데이터와 거짓 데이터를 잘 구분해내는 것이 목적입니다.

Generator와 Discriminator가 GAN 신경망 내에서 작동하는 방식과 절차는 다음과 같습니다.

1. Noise 확률 분포로부터 Noise샘플(z)을 추출하여 Generator의 입력값으로 넣습니다.

2. Generator는 이전 단계에서 입력된 Noise샘플을 내부 신경망을 거치게 하여 새로운 벡터값(G(z))을 생성하고 출력합니다.

3. 실측 데이터 분포로부터 data 샘플(x)를 추출합니다.

4. Discriminator는 이전 단계에서 추출한 샘플 x와 벡터값 G(z)를 입력으로 받습니다.

5. Discriminator는 입력된 x와 G(z)를 내부 신경망을 거치게 하여, G(z)가 실측데이터 분포로부터 온것인지(1) 생성모델로부터 온것인지(0) 예측(판별)하는 확률 값을 return 합니다.

따라서, Generator가 거짓 데이터를 생성하면 discriminator는 실제 데이터와 generator가 생성한 거짓 데이터로부터 실제와 거짓을 구분합니다. Generator는 discriminator를 속이지 못한 정보를, discriminator는 generator로 인해 속은 정보를 입력 받아 학습하고 이 과정을 반복합니다. 학습이 반복되면 generator는 점점 실제와 가까운 데이터를 생성하게 되고, discriminator는 점점 데이터의 실제와 거짓을 잘 구분하게 됩니다. 충분히 학습된 generator를 이미지 생성에 이용하게 되는 것입니다.

문제 2. 아래 GAN 코드를 읽으면서 다이어그램의 각 박스, 다이아몬드, 화살표에 대응되는 부분에 대해서 설명하는 코멘트를 추가한다. 예를 들어, " ## 다음 함수 호출은 새로운 fake image 를 생성하는데 사용될 랜덤넘버 z 를 생성한다" 라고 쓰면 된다.  

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, ), (0.5, )),
])
```

```python
batch_size = 128
z_dim = 100
```

```python
train_loader = torch.utils.data.DataLoader(
    dataset.MNIST('mnist', train = True, download = True, transform = transform),
    batch_size = batch_size,
    shuffle = True
)
```

```python
train_loader
```

```python
## 다음 함수는 새로운 fake image 를 생성하는데 사용될 랜덤넘버 z 를 생성하기 위해 작성되었다.

def weights_init(m):
    classname = m.__class__.__name__
    if classname.find('Conv') != -1:
        nn.init.normal_(m.weight.data, 0.0, 0.02)  # mean =0, sd = 0.02
    elif classname.find('BatchNorm') != -1:
        nn.init.normal_(m.weight.data, 1.0, 0.02)
        nn.init.constant_(m.bias.data, 0)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
## 해당 클래스(Generator)는 랜덤 벡터 z를 입력으로 받아 가짜 이미지를 출력한다.

class Generator_model(nn.Module):
    def __init__(self, z_dim):
        super().__init__()
        self.fc = nn.Linear(z_dim, 256 * 7 * 7)  ## 문제 3a.  self.fc은 input vector의 size가 z_dim (100) 이고, output vector 의 size 가 256*7*7 인 
                                                 ## fully connected network 이다. 256*7*7  = 12544 인데, 12544 라고 표시하지 않고, 256*7*7 이라고 표기한 
                                                 ## 프로그래머의 의도는 무엇인가? 
        self.gen = nn.Sequential(
            nn.ConvTranspose2d(256, 128, 4, 2, 1),  ## 문제 3b. (256, 128) 은 무엇을 의미하는가? 3번째 파라미터 4 는 무엇을 의미하는가?
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.01),
            nn.ConvTranspose2d(128, 64, 3, 1, 1), 
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.01),
            nn.ConvTranspose2d(64, 1, 4, 2, 1), 
            nn.Tanh()
        )
    def forward(self, input):
        x = self.fc(input)
        x = x.view(-1, 256, 7, 7)
        return self.gen(x)
```

```python
## 생성된 Generator 모델은 이전 단계에서 입력된 Noise샘플을 내부 신경망을 거치게 하여 새로운 벡터값(G(z))을 생성하고 출력합니다.

generator = Generator_model(z_dim).to(device)
```

```python
generator.apply(weights_init)
```

문제 4. 일반적으로 컨벌루션 신경망은 해상도가 큰 이미지에서 특징을 추출하면서 해상도가 적은 이미지들을 생산한다. generator 망은 "de-convolution" 망이라고 불린는 것인데, 어떤 점에 de-convolution  인가? 

문제 4 답) Generator망은 랜덤 벡터로 부터 'deconvolution'을 이용하여 특징으로부터 이미지를 확장해나간다는 점에서 CNN을 활용하는 Descriminator망이 이미지에 대해 real과 fake를 labelling하면서 이미지로 부터 특징을 추출하는 'convolution'을 활용하는것과 상반되는 방식을 가지고 있다.

```python
summary(generator, (100, ))
```

```python
## 해당 클래스(Discriminator)는 이전 단계에서 추출한 샘플 x와 벡터값 G(z)를 입력으로 받아 이미지가 진짜인지 가짜인지 출력합니다.

class Discriminator_model(nn.Module):
    def __init__(self):
        super().__init__()
        self.disc = nn.Sequential(
            nn.Conv2d(1, 32, 3, 2, 1), ##  1= in_channels [28x28x1], 32 = out_channels [14x14x32], 3 = kernel size, 2= stride, 1 = padding
            nn.LeakyReLU(0.01),
            nn.Conv2d(32, 64, 3, 2, 1),  ## 14*14*32 ==> 7*7*64
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.01),
            nn.Conv2d(64, 128, 3, 2, 1), ##  7*7*64 =>  4 *4 * 128 = 2048
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.01)
        )
        self.fc = nn.Linear(2048, 1)
    def forward(self, input):  # input = input image
        x = self.disc(input)
        return F.sigmoid(self.fc(x.view(-1, 2048)))
```

```python
## 생성된 Discriminator는 입력된 x와 G(z)를 내부 신경망을 거치게 하여, G(z)가 실측데이터 분포로부터 온것인지(1) 생성모델로부터 온것인지(0) 예측(판별)하는 확률 값을 return 합니다.
discriminator = Discriminator_model().to(device)
```

```python
discriminator.apply(weights_init)
```

문제 5. discriminator 신경망은 인풋 이미지를 분류를 하는 것을 목적으로 하는 신경망인데, 여기서는 인풋 이미지에 대해 어떤 판단을 내리는 분류기인가?  

문제 5 답) Discrimninator 신경망은 인풋 이미지에 대해 그 이미지가 진짜일 확률을 0과 1 사이의 숫자 하나로 판단하여 return하는 분류기이다.

```python
summary(discriminator, (1, 28, 28))
```

```python
criterion = nn.BCELoss()
```

문제 6. 위의 criterion 은 GAN에서 사용하는 판단기준인데, 이것은 어떤 기준인가?

문제 6 답) BCELoss는 기존의 cross entropy 손실 함수의 한 케이스로 이진 분류에 특화 되어있기 때문에 GAN에서 사용한다.

```python
# create a batch (whose size 64) of fixed noise vectors (z_dim=100)
fixed_noise = torch.randn(64, z_dim, device=device) # fixed_noise = a tensor (matrix) of 64 x 100 random numbers from Gaussian normal distribution with 0 mean and standard deviation 1
```

문제 7. 위의 fixed_noise 는 자료 구조 (벡터, 매트릭스..) 는 무엇이며, GAN에서 어떤 용도로 쓰이는가?

문제 7 답) fixed_noise의 자료 구조는 tensor(텐서)이며 Generative의 입력값으로  Noise 확률 분포로부터 추출된 Noise 샘플을 담기위해 사용된다.

```python
doptimizer = optim.Adam(discriminator.parameters())
goptimizer = optim.Adam(generator.parameters())
```

```python
real_label, fake_label = 1, 0
```

```python
image_list = []

g_losses = []
d_losses = []
iterations = 0

num_epochs = 50
```

```python
for epoch in range(num_epochs):
    print(f'Epoch : | {epoch+1:03} / {num_epochs:03} |')
    for i, data in enumerate(train_loader):

        # process the discriminator  network
        
        discriminator.zero_grad()
        
        real_images = data[0].to(device)
      
        size = real_images.size(0)
        # print ("for each batch, size =" + size)

        label = torch.full((size,), real_label, device=device)  ## 문제 8. 이 문장이 하는 일은 무엇인가?

                                                                ## 문제 8 답) 1로 구성된 image size와 같은 크기의 tensor를 return합니다.

        # print ("for each batch, label =" + label)

        d_output = discriminator(real_images).view(-1)

        # print ("for each batch, d_output (real) =" + d_output)

        derror_real = criterion(d_output, label) # 문제 9. 이 문장에서는 d_output 과 label 을 비교하여 무엇을 계산하는가? 그리고 이 계산을 하는 이유는 무엇인가?
                                                 # 문제 9 답) Binary cross entropy을 이용해 이미지가 진짜일 확률이 얼마나 정답과 가까운지를 측정해 학습 모델을 평가하기 위해 사용합니다.

        # print ("for each batch, derror_real =" + derror_real)

        derror_real.backward()  # 문제 10: 이 명령은 무엇을 하는가?
                                # 문제 10 답) derror_real의 도함수를 계산합니다.
        
        noise = torch.randn(size, z_dim, device=device)

        fake_images = generator(noise)

        label.fill_(0)  #_: in-place-operation  문제 11: 이것은 label 벡터에 0 을 채워넣는 작업인데, 이는 모든 label 값을 0 으로 설정한다는 말이다.
                        # 이것은 어떤 정보를 표현하는 것인가?               
                        # 문제 11 답) Generative 모델로 부터 왔음을 나타내는 확률 값을 표현합니다.

        d_output = discriminator(fake_images.detach()).view(-1)

        #print ("for each batch, d_output(fake) =" + d_output)

        derror_fake = criterion(d_output, label) 

        #print ("for each batch, derror_fake =" + derror_fake)
        derror_fake.backward()
        
        derror_total = derror_real + derror_fake  
              # 문제 12: derror_real 과 derror_fake 는 현재 학습중인 discriminator 망의 성능을 나타내는 지표인데 각각 무엇을 나타내는가?  이 둘을 합하여 전체 error 를 구하는 이유는 무엇인가?
              # 문제 12 답) derror_real 실측데이터 분포로부터 온것인지에 대한 지표이고 derror_fake는 Generator 모델로부터 온것인지에 대한 지표입니다. 
              #             Cross-Entropy Error를 활용하기 때문에 이 둘을 합하여 전체 error를 구하게 됩니다.

        doptimizer.step()  # update the discriminator weights

        # After the discriminator network is processed, now process the generator network
        # 문제 13. Generator network 의 목표는 무엇인가? discriminator network 과 연관지어 기술하라.
        # 문제 13 답) Generator는 실제 데이터를 학습해 그와 비슷한 거짓 데이터를 생성하면서 최대한 실제 데이터와 비슷한 거짓 데이터를 생성하는 것이 목적입니다.
        generator.zero_grad()

        # The generator's loss quantifies how well it was able to trick the discriminator. 
        # Intuitively, if the generator is performing well, the discriminator will classify the fake images as real (or 1). 
        # So we will compare the discriminators decisions on the generated images to an array of 1s.
       

        #   # 문제 14. gerror 는 generator 망의 성능을 평가하는 기준인데, 이를 계산하는 과정에서 generator 망을 직접 언급하지 
        # 않고 discrimator 를 언급한다. 이렇게 해도 gerror를 계산할 수 있는 이유를 설명하라.

        ## 문제 14 답) generator는 discrimator 극대화하고 싶어하는 것을 극소화하고 싶어하기 때문에 generator 망을 직접 언급하지 않고 discrimator를 언급하는 것이 가능합니다.
        
        label.fill_(real_images) #_: in-place-operation; the same as label.fill_(1)

        d_output = discriminator(fake_images).view(-1)

        gerror = criterion(d_output, label) #Cross-entropy is a measure of the difference between two probability distributions

        gerror.backward()  # compute the gradient of the objective function with respect to weight vector
        
        goptimizer.step()   # update the generator weights
        
        if i % 50 == 0:
            print(f'| {i:03} / {len(train_loader):03} | G Loss: {gerror.item():.3f} | D Loss: {derror_total.item():.3f} |')
            g_losses.append(gerror.item())
            d_losses.append(derror_total.item())
            
        if (iterations % 500 == 0) or ((epoch == num_epochs-1) and (i == len(train_loader)-1)):
            with torch.no_grad():
                fake_images = generator(fixed_noise).detach().cpu()

            image_list.append(vutils.make_grid(fake_images, padding=2, normalize=True))
            
        iterations += 1
```

```python
plt.figure(figsize=(10,5))
plt.title("Generator and Discriminator Loss During Training")
plt.plot(g_losses,label="Generator")
plt.plot(d_losses,label="Discriminator")
plt.xlabel("iterations")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
